### Шаг 2. Предобработка данных

Моя цель определять рыночную стоимость объектов недвижимости. Так что недвижимость без местоположения я удаляю.

```python
data = data.dropna(subset =['locality_name']).reset_index(drop=True)

data[(data['ceiling_height'] < 2)]
# Здесь всего 3 строчки с потолком ниже 2х метров. Они сильно не повлияют на данные так что я их оставляю
# Возможно здесь была опечатка вместо 2 впечатали 1.
```

```python
data.query('living_area < 4') # Здесь я вывел жилую площадь меньше 4м. Видно что это,возможно, опечатка т.к есть больщая кухня
# и площадь квартиры. Данные оставляю.
```

```python
def change_nan(column, what):
    """
    Меняет Nan . column - в какой колонке меняю. what на что меняю
    """
    data[column] = data[column].fillna(what)
    
change_nan('balcony', 0)
```

Mожно предположить логичную замену. Что человек не указал количество балконов — скорее всего, их нет. Такие пропуски заменим на 0.

```python
data['balcony'] = data['balcony'].astype(int) # Сразу ставлю целочисленное т.к 1.5 балкона быть не может
```

Аномальных значений потолка, которые меньше 2м. всего 3 шт. Я заполню отсутствующие пропуски медианой которая составляет 2.65
(что похоже на правду)

```python
change_nan('ceiling_height', data['ceiling_height'].median())
```

Заменю пропущенные значения на 0 т.к возможно это объявление еще висит и его еще не снимали.

```python
change_nan('days_exposition', 0)
data['days_exposition'] = data['days_exposition'].astype(int)
```

Меняю пропущенные значения в парках и водоёмах на 0 и ставлю целочисленное. Возможно их нет рядом или нет информации о них.
Возможно, просто не собирали информацию т.к. не всем они нужны.

```python
change_nan('parks_around3000',0)
data['parks_around3000'] = data['parks_around3000'].astype(int)
change_nan('ponds_around3000',0)
data['ponds_around3000'] = data['ponds_around3000'].astype(int)
```

```python
data['last_price'] = data['last_price'].astype(int)
# Округляю цену на снятия до Int т.к цифры после "." сильную роль не сыграют.
```

```python
# До перехода в нижний регистр в столбце locality_name было 364 уникальных значения. После перехода ничего не изменилось.
#Лучше оставлю так.

#Была еще гипотеза что все студии это апартаменты, но так как их меньше 0.1% не стал менять значения 
#т.к они не повлияют существенно на данные

data['is_apartment'].value_counts()# Посчитаем кол-во булевых значений и видим что апартаментов очень мало.
# Я предполагаю что все пропущенные значения это не апартаменты.
```

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
# Меняю все пропущенные значения на False из-за своей гипотезы

data.duplicated().sum() # Видно что нет дубликатов.
```

### Шаг 3. Посчитаю и добавлю в таблицу

```python
#Начнем с даты
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d' )
#Изменил тип на дату,округлил до дня т.к. час нас не интересует(Нам нужен день публикации)

data['day'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
# Сделал новые столбцы в какой день, месяц и год выставили объявление
```

```python
def group_floor(row): # Я думаю где пропущенно значение в floors_total(кол-во этажей) то floor(этаж) это последний.
    """
    Функция для группировки этажей на 4 группы первый,другой,последний и дом(Одноэтажная постройка).
    
    """
    try:
        floor = row['floor']
        total = row['floors_total']
        if floor == 1:
            if total > 1:
                return 'Первый'
            else:
                return 'Дом'
        if floor == total or total < 1:
            return "Последний"
        return 'Другой'
    except:
        return 'Ошибка!'


data['category_floor'] = data[['floor', 'floors_total']].apply(group_floor, axis = 1) # Сделал группировку этажей.

data['cost_for_metr'] = data['last_price'] / data['total_area']

data['cost_for_metr'] = data['cost_for_metr'].astype(int) # добавил в таблицу: цена квадратного метра и перевел в инт.

data['ratio_living_area'] = data['living_area'] / data['total_area'] 
data['ratio_kitchen'] = data['kitchen_area'] / data['total_area']
#Соотношение жилой и общей площади, а также отношение площади кухни к общей.

data.locality_name.value_counts()
```

Видно что кроме Санкт - Петербурга данные из других городов незначительные.Так что сделаю новый столбец с топ 10 
и остальную часть сгруппирую как "Область"

```python
top_ten = data.locality_name.value_counts().head(10).index # Переменная с топ 10 локациями для группировки

data['group_locality'] = data['locality_name'].where(data['locality_name'].isin(top_ten), 'Область') # группировка

data.sample(5) #Смотрю что теперь есть по 5 случайным строкам.
```

В *living_area* и *kitchen_area* есть пропущенные значения, которые я не менял. Возможно там свободная планировка или студия. Из-за этого я не стал менять значения. Следовательно, в *ratio_living_area* и *ratio_kitchen* тоже присутствуют значения Nan. В *data['day']* счет идет с 0 до 6, то есть 0 это понедельник. В столбце *group_locality* есть 10 населённых пунктов и одна группировка которая объединяет в себе более мелкие данные в группу "Область".

От дня выставления объявления слегка зависит цена, возможно кто продаёт не в настроении и решил слегка завысить цену.Соответственно цена зависит от расположения. Чем ближе к центру тем дороже. Есть районы, комплексы которые не уступают, а то и дороже квартир в центре это связано скорее всего с хорошей инфраструктурой, новизной постройки, качеством самого дома и т.д. Так же я выяснил что люди которые покупают квартиру в центре предпочитают "последний" и "Другой" этаж. Вне центра же предпочитают "последний",кстати первый в центре тоже не любят.

Можно предположить что мошенники повышают цены на имеющиеся объявления путем выставления его снова, но с ценой повыше. Это видно из графиков дней недели т.к. после снижение цены идет резкий рост в след. день. Это хорошо видно по графикам размещения по дням а центре Питера и вне его. Так что для выявления  аномалии или мошенническую деятельность лучше всего смотреть по  дням недели особо по будним дням и соответственно сравнить по более точным параметрам таких как адрес и номер квартиры.
